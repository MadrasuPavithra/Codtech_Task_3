```python
# 📦 Install required libraries (run only once)
!pip install jupyter-dash plotly pandas --quiet

# ✅ Import Libraries
import pandas as pd
from jupyter_dash import JupyterDash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px
import plotly.graph_objects as go
from google.colab import files
import io

# 📁 Upload your dataset (upload will pop up)
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
df = pd.read_csv(io.BytesIO(uploaded[file_name]), encoding='ISO-8859-1')

# ✅ Check required columns
required = {'Region', 'Category', 'Sales', 'Profit'}
if not required.issubset(df.columns):
    raise ValueError(f"Dataset must contain these columns: {required}")

# 🚀 Initialize Dash app
app = JupyterDash(__name__)
app.title = "Global Superstore Dashboard"

# 🎨 App layout
app.layout = html.Div([
    html.H1("📊 Global Superstore Dashboard", style={'textAlign': 'center'}),

    dcc.Dropdown(
        id='region-dropdown',
        options=[{'label': r, 'value': r} for r in sorted(df['Region'].dropna().unique())],
        placeholder="Select a Region (Optional)",
        value=None,
        style={'width': '50%', 'margin': 'auto'}
    ),

    dcc.Graph(id='graph1'),
    dcc.Graph(id='graph2'),
    dcc.Graph(id='graph3'),
    dcc.Graph(id='graph4'),
    dcc.Graph(id='graph5'),
    dcc.Graph(id='graph6'),
    dcc.Graph(id='graph7'),
    dcc.Graph(id='graph8'),
])

# 🔁 Callback to update all graphs
@app.callback(
    [Output(f'graph{i}', 'figure') for i in range(1, 9)],
    [Input('region-dropdown', 'value')]
)
def update_graphs(selected_region):
    filtered = df[df['Region'] == selected_region] if selected_region else df
    region_label = selected_region if selected_region else "All Regions"

    # Graph 1: Sales by Category (Bar)
    fig1 = px.bar(filtered, x='Category', y='Sales', color='Category', title='Sales by Category')

    # Graph 2: Profit by Category (Line)
    fig2 = px.line(filtered.groupby('Category')['Profit'].sum().reset_index(),
                   x='Category', y='Profit', title='Profit by Category', markers=True)

    # Graph 3: Sales Share by Category (Pie - region-specific)
    pie_data = filtered.groupby('Category')['Sales'].sum().reset_index()
    fig3 = px.pie(pie_data, names='Category', values='Sales',
                  title=f"Category Sales Share in {region_label}", hole=0.3)

    # Graph 4: Sales Gauge
    fig4 = go.Figure(go.Indicator(
        mode="gauge+number",
        value=filtered['Sales'].sum(),
        title={'text': f"Total Sales in {region_label}"},
        gauge={'axis': {'range': [0, df['Sales'].max()]}}
    ))

    # Graph 5: Box Plot of Profit by Category
    fig5 = px.box(filtered, x='Category', y='Profit', color='Category',
                  title='Profit Distribution by Category')

    # Graph 6: Stacked Sales by Region & Category (global view)
    grouped = df.groupby(['Region', 'Category'])['Sales'].sum().reset_index()
    fig6 = px.bar(grouped, x='Region', y='Sales', color='Category', barmode='stack',
                  title='Stacked Sales by Region & Category')

    # Graph 7: Scatter Plot of Sales vs Profit
    fig7 = px.scatter(filtered, x='Sales', y='Profit', color='Category',
                      title='Sales vs Profit by Category', size='Sales',
                      hover_data=['Region'])

    # Graph 8: Segment-Wise Sales (if Segment column exists)
    if 'Segment' in filtered.columns:
        fig8 = px.bar(filtered.groupby('Segment')['Sales'].sum().reset_index(),
                      x='Segment', y='Sales', color='Segment',
                      title=f"Segment-wise Sales in {region_label}")
    else:
        fig8 = go.Figure()
        fig8.add_annotation(text="Segment data not available", showarrow=False)

    return fig1, fig2, fig3, fig4, fig5, fig6, fig7, fig8

# ▶️ Run the app
app.run(debug=False)
```
